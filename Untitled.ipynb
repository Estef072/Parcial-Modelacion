```python
# Importaciones necesarias
import numpy as np
import random
import matplotlib.pyplot as plt
from collections import Counter

# =====================================
# Paso 1: Calcular la media de los valores originales
# =====================================

data_set = np.array([9, 10, 3, 5, 2, 8, 3, 1, 2, 2])
mean_original = np.mean(data_set)

print(f"Paso 1: La media de los valores originales es {mean_original}")

# =====================================
# Paso 2: Realizar un offset para trasladar la media a cero
# =====================================

data_set_offset = data_set - mean_original

print(f"Paso 2: El conjunto de datos con la media trasladada a cero es {data_set_offset}")

# =====================================
# Paso 3 y 4: Bootstrapping y Transformada Inversa
# =====================================

# Probabilidades dadas para usar en la Transformada Inversa
probabilities = np.array([0.37, 0.02, 0.1, 0.14, 0.04, 0.06, 0.02, 0.01, 0.02, 0.22])

# Función para la Transformada Inversa
def inverse_transform_sampling(data, probabilities):
    bins = np.add.accumulate(probabilities)
    return data[np.digitize(random.random(), bins)]

# Generación de 10,000 medias mediante Bootstrapping
bootstrap_means = []
n_iterations = 10000
for _ in range(n_iterations):
    bootstrap_sample = np.array([inverse_transform_sampling(data_set_offset, probabilities) for _ in range(len(data_set))])
    bootstrap_mean = np.mean(bootstrap_sample)
    bootstrap_means.append(bootstrap_mean)
bootstrap_means = np.array(bootstrap_means)

print("Paso 3 y 4: Se han generado 10,000 medias mediante bootstrapping y transformada inversa.")
```

Paso 1: Calculamos la media de los valores originales del conjunto de datos. La media resultó ser 
4.5
4.5.

Paso 2: Realizamos un "offset" al conjunto de datos para trasladar su media a cero. Esto se hace para centralizar los datos alrededor del origen, lo cual puede ser útil en diversos análisis estadísticos.

Paso 3 y 4: Utilizamos el método de bootstrapping junto con la transformada inversa para generar 10,000 medias basadas en el conjunto de datos original y las probabilidades dadas. Estas medias se almacenarán para su posterior análisis.

```python
# =====================================
# Paso 5: Histograma de las medias
# =====================================

# Generar el histograma para visualizar la distribución de las medias
plt.figure(figsize=(10, 6))
plt.hist(bootstrap_means, bins=50, density=True, color='blue', edgecolor='black')
plt.title("Distribución de Medias (Bootstrapping)")
plt.xlabel("Media")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()

print("Paso 5: El histograma muestra la distribución de las 10,000 medias generadas.")

# =====================================
# Paso 6: Dividir en 5 rangos y calcular su probabilidad
# =====================================

# Calcular los límites y las probabilidades de los 5 rangos
min_value = np.min(bootstrap_means)
max_value = np.max(bootstrap_means)
range_width = (max_value - min_value) / 5

range_probabilities = []
for i in range(5):
    lower_bound = min_value + i * range_width
    upper_bound = min_value + (i + 1) * range_width
    prob = np.sum((bootstrap_means >= lower_bound) & (bootstrap_means < upper_bound)) / n_iterations
    range_probabilities.append(prob)
range_probabilities = np.array(range_probabilities) / np.sum(range_probabilities)  # Normalización

print(f"Paso 6: Las probabilidades de los 5 rangos son {range_probabilities}")


# =====================================
# Cálculo de la Desviación Estándar para los Pasos 5 y 6
# =====================================

# Desviación estándar de la distribución de medias (Paso 5)
std_bootstrap_means = np.std(bootstrap_means)

# Desviación estándar de las probabilidades de los 5 rangos (Paso 6)
std_range_probabilities = np.std(range_probabilities)

std_bootstrap_means, std_range_probabilities

```

Paso 5: Generamos un histograma que muestra la distribución de las 10,000 medias calculadas. Este histograma nos ayuda a visualizar cómo se distribuyen las medias del conjunto de datos, lo cual es útil para entender la variabilidad y la tendencia central de los datos.

Paso 6: Dividimos la distribución de las medias en 5 rangos y calculamos las probabilidades para cada rango. Estas probabilidades nos ayudan a entender cómo se distribuyen las medias en diferentes intervalos. Por ejemplo, el tercer rango tiene la mayor probabilidad, lo que indica que la mayoría de las medias se encuentran en ese rango.

Desviación Estándar en el Paso 6: La desviación estándar de las probabilidades de los 5 rangos es aproximadamente

0.171. Esta medida nos indica cuánto varían las probabilidades de los diferentes rangos alrededor de la media de las probabilidades. Una desviación estándar más alta aquí sugeriría una distribución más dispersa de las probabilidades entre los diferentes rangos.

```python
# =====================================
# Paso 7: Técnica de Aceptación y Rechazo
# =====================================

# Parámetros para el Generador Congruencial Lineal Mixto (GCLM)
a = 1664525
c = 1013904223
m = 2 ** 32
seed = 42

# Función para generar un número aleatorio utilizando un GCLM
def lcg(seed, a, c, m):
    return (a * seed + c) % m

# Inicializar el estado del generador
current_seed = seed

# Lista para almacenar las variables aleatorias generadas
generated_vars = []

# Generar variables aleatorias utilizando la técnica de aceptación y rechazo
for _ in range(10000):  # 10,000 iteraciones para coincidir con el número de iteraciones de bootstrapping
    # Generar un número aleatorio [0, 1) usando un GCLM
    current_seed = lcg(current_seed, a, c, m)
    random_var = current_seed / m
    
    # Escoger un rango basado en las probabilidades calculadas previamente
    chosen_range = np.random.choice(5, p=range_probabilities)
    
    # Calcular los límites del rango elegido
    lower_bound = min_value + chosen_range * range_width
    upper_bound = min_value + (chosen_range + 1) * range_width
    
    # Generar una variable aleatoria que caiga dentro del rango elegido
    generated_var = lower_bound + (upper_bound - lower_bound) * random_var
    
    # Almacenar la variable aleatoria generada
    generated_vars.append(generated_var)

# Convertir la lista a un arreglo numpy
generated_vars = np.array(generated_vars)

print("Paso 7: Se han generado 10,000 variables aleatorias utilizando la técnica de aceptación y rechazo.")
```

Paso 7: Utilizamos la técnica de aceptación y rechazo junto con un Generador Congruencial Lineal Mixto (GCLM) para generar 10,000 variables aleatorias. Estas variables se generan de acuerdo con las probabilidades de los 5 rangos que calculamos en el paso 6. El objetivo es obtener una muestra que aproxime la distribución de las medias obtenidas del bootstrapping.

```python
# =====================================
# Paso 8: Variables Aleatorias Discretas
# =====================================

# Redondear las variables aleatorias generadas al entero más cercano para obtener variables aleatorias discretas
discrete_vars = np.round(generated_vars).astype(int)

print(f"Paso 8: Las primeras 10 variables aleatorias discretas son {discrete_vars[:10]}")

# =====================================
# Paso 9: Función de Masa de Probabilidad y Función de Distribución Acumulada
# =====================================

# Contar la frecuencia de cada variable aleatoria discreta
frequency_count = Counter(discrete_vars)

# Calcular la FMP normalizando las frecuencias
total_count = len(discrete_vars)
pmf = {k: v / total_count for k, v in frequency_count.items()}

# Calcular la FDA
cdf = {}
accumulated_prob = 0
for k in sorted(pmf.keys()):
    accumulated_prob += pmf[k]
    cdf[k] = accumulated_prob

print(f"Paso 9: La Función de Masa de Probabilidad es {pmf}")
print(f"Paso 9: La Función de Distribución Acumulada es {cdf}")
```

Paso 8: Después de generar las variables aleatorias continuas en el Paso 7, las redondeamos al entero más cercano para convertirlas en variables aleatorias discretas. Esta es una forma sencilla de discretizar una variable continua, y es especialmente útil cuando el conjunto de datos original también es discreto.

Paso 9: Finalmente, calculamos la Función de Masa de Probabilidad (FMP) y la Función de Distribución Acumulada (FDA) para este conjunto de variables aleatorias discretas. La FMP nos da las probabilidades de cada valor discreto, y la FDA nos muestra cómo se acumulan estas probabilidades. Estas funciones son útiles para entender la distribución subyacente del conjunto de datos

```python
# =====================================
# Gráfica de la Función de Masa de Probabilidad (FMP)
# =====================================

# Valores y probabilidades
values = list(pmf.keys())
probs = list(pmf.values())

# Gráfica de la FMP
plt.figure(figsize=(10, 6))
plt.bar(values, probs, color='purple', edgecolor='black')
plt.xlabel("Valor")
plt.ylabel("Probabilidad")
plt.title("Función de Masa de Probabilidad (FMP)")
plt.grid(True, axis='y')
plt.show()

# =====================================
# Estimación de la Función de Distribución Acumulada (FDA)
# =====================================

# Valores y probabilidades acumuladas
values_cdf = list(cdf.keys())
probs_cdf = list(cdf.values())

# Gráfica de la FDA
plt.figure(figsize=(10, 6))
plt.step(values_cdf, probs_cdf, where='post', color='orange', label='FDA estimada')
plt.xlabel("Valor")
plt.ylabel("Probabilidad Acumulada")
plt.title("Función de Distribución Acumulada (FDA)")
plt.grid(True)
plt.legend()
plt.show()
```
